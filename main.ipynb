```python
import os
import glob

project_dir = "/workspace/ese-3060-project/"
original_data_source = os.path.join(project_dir, "fineweb10B")
data_dir = os.path.join(project_dir, "data", "fineweb10B")

# Create the target directory if it doesn't exist
os.makedirs(data_dir, exist_ok=True)

# Move fineweb_*.bin files from original_data_source to data_dir
bin_files = glob.glob(os.path.join(original_data_source, "fineweb_*.bin"))

if bin_files:
    print(f"Moving {len(bin_files)} .bin files from {original_data_source} to {data_dir}")
    for f in bin_files:
        os.rename(f, os.path.join(data_dir, os.path.basename(f)))
else:
    print(f"No fineweb_*.bin files found in {original_data_source} to move.")

# Verify the files are in the new location
print("Files in new data directory:")
print(os.listdir(data_dir))
```
